```python
# Imports

import os
from dotenv import load_dotenv

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

load_dotenv()

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY", "AIzaSyAfpHjaT9hrP0Ha1esiY9L2zzklWrPr3bc")

if not GOOGLE_API_KEY:
    raise RuntimeError("Missing GOOGLE_API_KEY in environment.")

print("Environment loaded successfully.")
```

```python
# Create the LLM

llm = ChatGoogleGenerativeAI(
    google_api_key=GOOGLE_API_KEY,
    model="gemini-2.5-flash",
    temperature=0.5,
    max_output_tokens=1000,
    convert_system_message_to_human=True,
)

print("LLM is ready.")
```

```python
# Very simple one-question chatbot

def ask_llm_once(user_question: str) -> str:
    messages = [
        SystemMessage(
            content=(
                "You are a helpful and clear AI assistant. "
                "Answer briefly and clearly."
            )
        ),
        HumanMessage(content=user_question),
    ]

    response = llm.invoke(messages)
    return response.content

# Example
question = "What is your favourite color?"
answer = ask_llm_once(question)

print("User:", question)
print("Bot:", answer)
```

```python
# Add chat history

chat_history = []

def ask_with_history(user_question: str, history: list) -> str:
    messages = [
        SystemMessage(
            content=(
                "You are a helpful AI assistant. "
                "Keep answers clear, short, and beginner-friendly."
            )
        )
    ]

    messages.extend(history)
    messages.append(HumanMessage(content=user_question))

    response = llm.invoke(messages)
    return response.content
```

```python
# Testing chat history - multi turn conversation

user_question_1 = "What is a chatbot backend then?"
bot_answer_1 = ask_with_history(user_question_1, chat_history)

chat_history.append(HumanMessage(content=user_question_1))
chat_history.append(AIMessage(content=bot_answer_1))

print("User:", user_question_1)
print("Bot:", bot_answer_1)
```

```python
user_question_2 = "How is it different from the frontend?"
bot_answer_2 = ask_with_history(user_question_2, chat_history)

chat_history.append(HumanMessage(content=user_question_2))
chat_history.append(AIMessage(content=bot_answer_2))

print("User:", user_question_2)
print("Bot:", bot_answer_2)
```
